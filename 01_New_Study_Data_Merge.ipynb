```python
import pandas as pd
# Veri setlerini okuma
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
weather_df = pd.read_csv('weather.csv')
holidays_df = pd.read_csv('holidays.csv')
```

```python
weather_df
```

```python
# Değişkenlerdeki eksik değer sayısı gözlemliyoruz.
weather_df.isnull().sum()
```

```python
# Değişkenlerdeki eksik değer sayısı gözlemliyoruz.
weather_df.notnull().sum()
```

```python
# Tarih ve saat bilgisini içeren sütunu seçin
date_time = weather_df['date']

# Boşluk karakterinden bölerek tarih ve saat bilgisini ayırın
weather_df['Date'] = date_time.str.split(' ').str[0]
weather_df['Time'] = date_time.str.split(' ').str[1]

# İlk sütunu tarih ve saat bilgisi olarak değiştirin
weather_df.drop(columns=['date'], inplace=True)
```

```python
weather_df
```

```python
# Tarih sütununu uygun formata dönüştürün
weather_df['Date'] = pd.to_datetime(weather_df['Date'])

# Eğitim veri setinizdeki son tarih
end_date_training = pd.to_datetime('2024-01-31')

# Son tarihten sonraki verileri filtreleyin
filtered_weather_df = weather_df[weather_df['Date'] <= end_date_training]
```

```python
filtered_weather_df
```

```python
filtered_weather_df.reset_index(drop=True, inplace=True)
filtered_weather_df
```

```python
weather_mean = filtered_weather_df.drop(columns=['Time', 'lat', 'lon'])
```

```python
new_weather_mean = weather_mean.groupby(['ilce', 'Date']).mean().reset_index()
new_weather_mean
```

```python
# 'tarih' sütununu uygun bir tarih/datetime formatına dönüştür
train_df['tarih'] = pd.to_datetime(train_df['tarih'])

# Eksik tarih aralığını oluştur
date_range = pd.date_range(start="2021-01-01", end="2024-01-31")

# Tüm ilçeleri ve tarihleri içeren bir DataFrame oluştur
all_dates = []
all_ilceler = []
for date in date_range:
    for ilce in train_df['ilce'].unique():
        all_dates.append(date)
        all_ilceler.append(ilce)
full_df = pd.DataFrame({'tarih': all_dates, 'ilce': all_ilceler})

# Train veri seti ile eksik tarihleri birleştir
train_merged_df = pd.merge(full_df, train_df, on=['tarih', 'ilce'], how='left')

# Eksik değerleri 0 olarak doldur
train_merged_df['bildirimli_sum'] = train_merged_df['bildirimli_sum'].fillna(0)
train_merged_df['bildirimsiz_sum'] = train_merged_df['bildirimsiz_sum'].fillna(0)
```

```python
train_merged_df
```

```python
# 'Date' ve 'tarih' sütunlarının isimlerini uyumlu hale getir
new_weather_mean = new_weather_mean.rename(columns={'Date': 'tarih'})

# 'ilce' sütunlarının isimlerini uyumlu hale getir
train_merged_df = train_merged_df.rename(columns={'district': 'ilce'})

# İki veri setini 'tarih' ve 'ilce' sütunlarına göre birleştir
merged_df = pd.merge(train_merged_df, new_weather_mean, on=['tarih', 'ilce'], how='left')
```

```python
merged_df
```

```python
# Veriseti içindeki "ilce" değişkenini dönüştürüyoruz.
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
lbe.fit_transform(merged_df["ilce"])

# Veriseti içine eklemek için yeni bir değişkene atıyoruz.
merged_df["new_ilce"] = lbe.fit_transform(merged_df["ilce"])
merged_df
```
